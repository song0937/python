```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

def get_beach_info(region):
    url = 'http://apis.data.go.kr/1192000/service/OceansBeachInfoService1/getOceansBeachInfo1'
    params = {
        'serviceKey': 'k3TNRfeuhP7Gi7+fYHrF4+E9GUpv/NNiEBCCIqiMaJZZFIDm4oQwxxbqRSBjdmNVvKhy9KXRamq55iEvWlx5/w==',
        'pageNo': '1',
        'numOfRows': '10',
        'SIDO_NM': region,
        'resultType': 'json'
    }

    response = requests.get(url, params=params)
    root = ET.fromstring(response.content)
    beach_data = []

    for beach in root.findall('.//body/items/item'):
        name = beach.findtext('staNm')  # 해수욕장 이름
        lat_text = beach.findtext('lat')  # 위도
        lon_text = beach.findtext('lon')  # 경도
        sido = beach.findtext('sidoNm')  # 시/도
        gungu = beach.findtext('gugunNm')  # 구/군
        link = beach.findtext('linkTel') # 연락처

        beach_data.append({
            '시/도': sido,
            '구/군': gungu,
            '해수욕장 이름': name,
            '연락처': link
        })

    return beach_data

def display_beach_info(beach_data):
    df = pd.DataFrame(beach_data)
    df = df[['시/도', '구/군', '해수욕장 이름', '연락처']]
    display(df)

def save_as_excel(beach_data, filename):
    df = pd.DataFrame(beach_data)
    df = df[['시/도', '구/군', '해수욕장 이름', '연락처']]
    df.to_excel(filename, index=False)

region = input('지역을 입력하세요: ')
beach_data = get_beach_info(region)
display_beach_info(beach_data)

filename = input('저장할 파일명을 입력하세요: ')
save_as_excel(beach_data, filename + '.xlsx')
```

```python
import requests
import folium
import xml.etree.ElementTree as ET

# 대한민국 지도 생성
map_korea = folium.Map(location=[36.5, 127.5], zoom_start=7)

sido_nm = input("찾고싶은 지역을 입력하세요: ")

url = 'http://apis.data.go.kr/1192000/service/OceansBeachInfoService1/getOceansBeachInfo1'
params = {
    'serviceKey': 'k3TNRfeuhP7Gi7+fYHrF4+E9GUpv/NNiEBCCIqiMaJZZFIDm4oQwxxbqRSBjdmNVvKhy9KXRamq55iEvWlx5/w==',
    'pageNo': '1',
    'numOfRows': '100',
    'resultType': 'json',
    'SIDO_NM': sido_nm
}

response = requests.get(url, params=params)
root = ET.fromstring(response.content)

beaches = root.findall('.//body/items/item')
if beaches:
    for beach in beaches:
        name = beach.findtext('staNm')  # 해수욕장 이름
        lat_text = beach.findtext('lat')  # 위도
        lon_text = beach.findtext('lon')  # 경도
        sido = beach.findtext('sidoNm')  # 시/도
        gungu = beach.findtext('gugunNm')  # 구/군
        link = beach.findtext('linkTel')  # 연락처

        if lat_text is None or lon_text is None:
            continue

        lat = float(lat_text)
        lon = float(lon_text)

        popup_text = f"<b>해수욕장 이름:</b> {name}<br>" \
                     f"<b>시/도:</b> {sido}<br>" \
                     f"<b>구/군:</b> {gungu}<br>" \
                     f"<b>연락처:</b> {link}"

        html = f"""
        <h3>{name}</h3>
        <p><b>시/도:</b> {sido}</p>
        <p><b>구/군:</b> {gungu}</p>
        <p><b>연락처:</b> {link}</p>
        """

        iframe = folium.IFrame(html=html, width=200, height=150)
        popup = folium.Popup(iframe, max_width=300)

        marker = folium.Marker(location=[lat, lon], popup=popup, tooltip=name)
        marker.add_to(map_korea)

    map_korea.save('map.html')
    print("마커가 지도에 추가되었습니다. 'map.html' 파일을 확인하세요.")

else:
    print("해수욕장 정보를 가져오는데 실패했습니다.")
```
